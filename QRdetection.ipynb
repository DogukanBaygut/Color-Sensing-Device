```python
import cv2
from google.colab.patches import cv2_imshow
import imutils

from ultralytics import YOLO
font=cv2.FONT_HERSHEY_TRIPLEX

img_path="/content/drive/MyDrive/Colab_Notebooks/runs/detect/predict2/IMG_0177_JPG.rf.3921d5b005b559bd4e6ed8420e3a0a82.jpg"
model_path="/content/drive/MyDrive/Colab_Notebooks/runs/detect/train/weights/best.pt"

model=YOLO(model_path)
img=cv2.imread(img_path)
img=imutils.resize(img,width=400)

results= model(img)[0]
threshold=0.5

for result in results.boxes.data.tolist():
  x1,y1,x2,y2,score,class_id=result
  x1,y1,x2,y2,class_id= int(x1),int(y1),int(x2),int(y2),int(class_id)

  if score>threshold:
    cv2.rectangle(img,(x1,y1),(x2,y2),(0,0,255),2)
    cv2.rectangle(img,(x1-50,y1-50),(x2+50,y2+50),(0,255,0),5)
    class_name=results.names[class_id]
    score=score*100

    text=f"{class_name}:Dogruluk orani: %{score:.2f}"
    cv2.putText(img,text,(x1,y1-10),font,0.5,(0,0,255),1,cv2.LINE_AA)

cv2_imshow(img)
```
